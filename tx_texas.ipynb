```python
import xarray as xr
from tqdm import tqdm

# Define the range of years and scenarios
years = range(2015, 2101)
scenarios = ['ssp585', 'ssp245']

# Define the latitude and longitude slice for Texas
lat_range = slice(25, 37)
lon_range = slice(253, 267)

# Create progress bar
total_iterations = len(years) * len(scenarios)
with tqdm(total=total_iterations, desc="Processing NetCDF files") as pbar:
    for year in years:
        for scenario in scenarios:
            # Construct the file path
            input_file = f'tc/tc_{scenario}_{year}.nc'
            output_file = f'tc_texas_{scenario}_{year}.nc'

            try:
                # Open the dataset and select the region
                ds = xr.open_mfdataset(input_file).sel(lat=lat_range, lon=lon_range)
                
                # Save the subsetted dataset
                ds.to_netcdf(output_file)
                tqdm.write(f"Processed {output_file}")

            except Exception as e:
                tqdm.write(f"Error processing {input_file}: {e}")

            # Update progress bar
            pbar.update(1)
```
